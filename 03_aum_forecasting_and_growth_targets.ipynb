# Business Objective
This analysis focuses on forecasting AUM growth and determining client acquisition targets for a mutual fund advisory. The objective is to model historical trends, evaluate growth patterns, and derive actionable targets to support long-term revenue planning.


```python
import pandas as pd

# Set initial values
cagr = 0.32  # 32%
roi = 0.15   # 15%
n_years = 10

# Starting AUM
initial_aum = 90912389

# Initialize list to store results
results = [{'Year': 2025, 'Target AUM': initial_aum, 'RoI on Prev AUM': None, 'New Inflow': None}]

# Generate projections for next 10 years
for i in range(1, n_years + 1):
    prev = results[-1]
    year = prev['Year'] + 1
    target_aum = prev['Target AUM'] * (1 + cagr)
    roi_aum = prev['Target AUM'] * (1 + roi)
    new_inflow = target_aum - roi_aum

    results.append({
        'Year': year,
        'Target AUM': round(target_aum),
        'RoI on Prev AUM': round(roi_aum),
        'New Inflow': round(new_inflow)
    })

# Convert results to DataFrame
df_targets = pd.DataFrame(results)
pd.set_option('display.float_format', '{:,.0f}'.format)
df_targets
```

```python
df_targets.to_excel("output.xlsx", index=False, float_format="%.0f")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# 1) Create DataFrame for the known (Clients, AUM) pairs
data_known = {
    'Clients': [49, 92, 127, 139, 165, 190, 214, 255],
    'AUM': [2568599.57, 17835113.21, 26234288.8, 27235648.18,
            39815070.75, 60286990.93, 76128503.72, 90912388.63]
}
df_known = pd.DataFrame(data_known)

# 2) Create DataFrame for the new AUM values
data_new = {
    'AUM': [120004353, 158405746, 209095585, 276006172, 364328147,
            480913154, 634805363, 837943079, 1106084864, 1460032020]
}
df_new = pd.DataFrame(data_new)

# 3) Train a linear regression model
X = df_known[['AUM']]  # Predictor
y = df_known['Clients']  # Target

model = LinearRegression()
model.fit(X, y)

# 4) Predict client counts for the new AUM values
X_new = df_new[['AUM']]
df_new['Predicted Clients'] = model.predict(X_new)

print(df_new)
```

```python
import numpy as np

def solve_for_clients(y, a=1020.7, b=137668, c_offset=7e6):
    """
    Solves the quadratic equation:
    1020.7 x^2 + 137668 x - (7e6 + y) = 0
    Returns the positive real solution for x (number of clients)
    """
    # Note: c_total = - (7e6 + y)
    c_total = - (c_offset + y)

    # Calculate discriminant
    discriminant = b**2 - 4 * a * c_total

    if discriminant < 0:
        print("No real solutions.")
        return None

    # Compute both roots
    x1 = (-b + np.sqrt(discriminant)) / (2 * a)
    x2 = (-b - np.sqrt(discriminant)) / (2 * a)

    # Choose the positive solution
    if x1 >= 0 and x2 >= 0:
        return min(x1, x2)  # you can choose the smaller positive root if both work
    elif x1 >= 0:
        return x1
    elif x2 >= 0:
        return x2
    else:
        print("No positive solution found.")
        return None

df_new['Predicted Clients (Poly)'] = df_new['AUM'].apply(lambda y: solve_for_clients(y, a=1020.7, b=137668, c_offset=7e6))
df_new
```
